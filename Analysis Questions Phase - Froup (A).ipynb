```python
import pyodbc
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Connect to SQL Server
conn = pyodbc.connect('DRIVER={SQL Server};'
                      'SERVER=AMRDESOKEY;'  # Replace with your server name
                      'DATABASE=SupplyChain;'  # Replace with your database name
                      'Trusted_Connection=yes;')
# Step 1: Define the SQL query
query1 = """
SELECT P.Product_type, SUM(S.Revenue_generated) AS Total_Revenue, I.Location
FROM Sales S
JOIN Products P ON S.SKU = P.SKU
JOIN Suppliers I ON I.SKU = P.SKU
GROUP BY P.Product_type, I.Location
ORDER BY Total_Revenue DESC;
"""

# Step 2: Execute the query and load the result into a Pandas DataFrame
df1 = pd.read_sql(query1, conn)

# Step 3: Close the connection
conn.close()

# Step 4: Display the DataFrame
print(df1.head())

# Step 5: Create a bar plot for Revenue by Product Type and Location
plt.figure(figsize=(12, 6))
sns.barplot(x='Product_type', y='Total_Revenue', hue='Location', data=df1)
plt.xticks(rotation=45)
plt.title('Revenue by Product Type and Location')
plt.ylabel('Total Revenue')
plt.xlabel('Product Type')
plt.tight_layout()
plt.show()
```

```python
import pyodbc
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Connect to SQL Server
conn = pyodbc.connect('DRIVER={SQL Server};'
                      'SERVER=AMRDESOKEY;'  # Replace with your server name
                      'DATABASE=SupplyChain;'  # Replace with your database name
                      'Trusted_Connection=yes;')
# Step 1: Define the SQL query
query3 = """
SELECT P.Product_type, I.Lead_times, SUM(S.Revenue_generated) AS Total_Revenue
FROM Sales S
JOIN Products P ON S.SKU = P.SKU
JOIN Inventory I ON I.SKU = P.SKU
GROUP BY P.Product_type, I.Lead_times
ORDER BY Total_Revenue DESC;
"""

# Step 2: Execute the query and load the result into a Pandas DataFrame
df3 = pd.read_sql(query3, conn)

# Step 3: Close the connection
conn.close()

# Step 4: Display the DataFrame
print(df3.head())

# Step 5: Create a scatter plot for Lead Times vs Revenue by Product Type
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Lead_times', y='Total_Revenue', hue='Product_type', data=df3)
plt.title('Lead Times vs. Revenue Generated by Product Type')
plt.ylabel('Total Revenue')
plt.xlabel('Lead Times (Days)')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Assuming you have a DataFrame named 'Clean_Data'

# Step 1: Map the Inspection results to numerical values
Clean_Data['Inspection results (numeric)'] = Clean_Data['Inspection results'].map({'Fail': 0, 'Pending': 1, 'Pass': 2})

# Step 2: Calculate the correlation coefficient
correlation = Clean_Data['Inspection results (numeric)'].corr(Clean_Data['Defect rates'])

# Define a function to label the correlation
def label_correlation(corr):
    if abs(corr) >= 0.8:  # Strong correlation
        label = "Strong " + ("Positive" if corr > 0 else "Negative") + " Correlation"
    elif abs(corr) >= 0.5:  # Moderate correlation
        label = "Moderate " + ("Positive" if corr > 0 else "Negative") + " Correlation"
    else:
        label = "Weak or No Correlation"
    return label

# Step 3: Create the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(Clean_Data['Inspection results (numeric)'],Clean_Data['Defect rates'], 
            c=Clean_Data['Inspection results (numeric)'], cmap='viridis')  # Use a colormap for better distinction
plt.xlabel('Inspection results')
plt.ylabel('Defect rates')
plt.title('Correlation between Inspection Results and Defect Rates')
plt.grid(True)

# Set custom tick labels for the x-axis to show original categories
plt.xticks([0, 1, 2], ['Fail', 'Pending', 'Pass'])

# Step 4: Add correlation label
corr_label = label_correlation(correlation)
plt.annotate(corr_label, xy=(0.1, 0.9), xycoords='axes fraction', fontsize=12, color='red')  # Adjust position and color

# Step 5: Display the plot
plt.show()
```
